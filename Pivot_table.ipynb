```python
import pandas as pd
from openpyxl import load_workbook
```

```python
df = pd.read_csv('Walmart_Sales.csv')
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, format='%d-%m-%Y')
```

```python
month = df["Date"].dt.month
df_pivot = df.copy()
df_pivot['Month'] = month
```

```python
pivot_table= df_pivot.pivot_table(index='Store', 
                                   columns='Holiday_Flag', 
                                   values='CPI', 
                                   aggfunc='sum').round(1)
pivot_table.sort_values(by=1, ascending=False)
```

```python
pivot_table.to_excel('Report.xlsx', sheet_name='WeeklySales')
```

```python
wb = load_workbook('Report.xlsx')
sheet = wb['WeeklySales']
```
